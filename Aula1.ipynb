```python
import pandas as pd

path_cli = "/content/aula1/clientes.csv"
path_ag =  "/content/aula1/agencias.csv"
path_output = "/content/aula1/cli_ag.csv"

cli_data_frame = pd.read_csv(path_cli)
ag_data_frame = pd.read_csv(path_ag)

cli_data_frame.head(10)

```

```python
ag_data_frame.head(5)
```

```python
#Juntango (merge) os 2 data frames e depois escrevendo em um arquivo csv
cli_ag_df = pd.merge(cli_data_frame, ag_data_frame, 
                              on='num_agencia', how='inner')
cli_ag_df.head(20)

cli_ag_df.to_csv(path_output,index=False)
```
